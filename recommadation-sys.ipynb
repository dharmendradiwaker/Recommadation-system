## **Popularity Based System**

#### Most Popular Books based on their Avg. Rating

```python
# Merge the Books dataset with rating dataset
ratings_with_name = rating.merge(books, on="ISBN")
```

```python
# checking the new dataset
ratings_with_name
```

```python
# Count the every book's number rating
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Number of rating'}, inplace=True)
```

```python
#checking the dataset
num_rating_df
```

```python
# drive the average rating of each book
avg_rating_df = ratings_with_name.groupby('Book-Title')
avg_rating_df = avg_rating_df['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
```

```python
# checking the dataset
avg_rating_df
```

```python
# Merge the data set of number of rating and average rating
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
#checking the dataset
popular_df
```

```python
# containing books where ratings are more than 250.
popular_df= popular_df[popular_df['Number of rating']>=250].sort_values('avg_ratings', ascending= False)
```

```python
popular_df.head(2)
```

```python
# Remove the duplicate data
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
popular_df
```

```python
# droping the duplicate values and choice usefull data columns.
popular_df = popular_df.drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Number of rating','avg_ratings']]
```

```python
# Top 20 Books with high number of average ratings.
popular_df[:20]
```

## **Collaborative Filtering Based Recommender System**

Move toward creating a Recommendation system.

```python
# Getting the users those gave rating to more than 200 books(Serious readers).
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users = x[x].index
```

```python
# checking the readers in rating_with_name dataset.

filter_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
# getting famous book which got rating more than 50.
y = filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = y[y].index
```

```python
# checking famous book is in the filter data or not.
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_book)]
```

```python
# creating a pivot table
pt=final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
# filling Nan values with 0
pt.fillna(0, inplace=True)
```

```python
# Checking the dataset
pt
```

```python
# Importing sklearn libaray module.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# creating similarity score
similarity_scores = cosine_similarity(pt)
```

```python
# checking similarity score
similarity_scores.shape
```

```python
#creating a function to get the recommadation.
def recommend(book_name):
  index = np.where(pt.index==book_name)[0][0]
  similar_item = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
  for i in similar_item:
    print(pt.index[i[0]])
```

```python
# Let's see the recommadation
recommend("Animal Farm")
```
